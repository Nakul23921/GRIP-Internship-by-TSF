In this task, we will predict the optimum number of clusters using K-Means Clustering and present it visually from the given 'iris' dataset.

```python
#Importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = pd.read_csv('Iris.csv', index_col = 0)

iris.head(50)
```

```python
iris.describe()
```

```python
iris.info()
```

# Finding the optimum number of clusters for k-means classification : 

```python
X = iris.drop('Species', axis = 1).values
```

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 11) :
  kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                  max_iter = 300, n_init = 10, random_state = 0)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # Within Cluster Sum of Squares
plt.show
```

# Applying K_Means to the Dataset

```python
# Choosing number of clusters as 3

kmeans= KMeans(n_clusters = 3, init = 'k-means++', max_iter = 500, n_init = 10, random_state = 0)

y_pred=kmeans.fit_predict(X)
```

```python
y_pred
```

# Visualizing the clusters - On the first two columns 

```python
plt.figure(figsize = (10, 5))
plt.scatter(X[y_pred == 0,1], X[y_pred == 0,2], s = 100, c = 'red', label = 'Iris-Setosa')
plt.scatter(X[y_pred == 1,1], X[y_pred == 1,2],s = 100,c = 'blue', label = 'Iris-Vercicolor')
plt.scatter(X[y_pred == 2,1], X[y_pred== 2,2], s = 100, c = 'green', label = 'Iris-Setosa')


# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:,1], kmeans.cluster_centers_[:,2], s = 100, c = 'black', label = 'centroids')
plt.legend()
```
